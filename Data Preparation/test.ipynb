```python
import vtk

# Path to your NRRD file
filename = "../segmentations_extraites/Threshold-200-MAX_label1.nrrd"

# Step 1: Read the NRRD file
reader = vtk.vtkNrrdReader()
reader.SetFileName(filename)
reader.Update()

# Step 2: Create a color and opacity transfer function
# (Useful if it's a label map or grayscale image)
colorFunc = vtk.vtkColorTransferFunction()
opacityFunc = vtk.vtkPiecewiseFunction()

# For simple label visualization (bone segmentation)
colorFunc.AddRGBPoint(0, 0.0, 0.0, 0.0)  # background
colorFunc.AddRGBPoint(1, 1.0, 1.0, 0.8)  # bone color

opacityFunc.AddPoint(0, 0.0)
opacityFunc.AddPoint(1, 1.0)

# Step 3: Volume properties
volumeProperty = vtk.vtkVolumeProperty()
volumeProperty.SetColor(colorFunc)
volumeProperty.SetScalarOpacity(opacityFunc)
volumeProperty.ShadeOn()
volumeProperty.SetInterpolationTypeToLinear()

# Step 4: Mapper
volumeMapper = vtk.vtkSmartVolumeMapper()
volumeMapper.SetInputConnection(reader.GetOutputPort())

# Step 5: Volume actor
volume = vtk.vtkVolume()
volume.SetMapper(volumeMapper)
volume.SetProperty(volumeProperty)

# Step 6: Renderer, Render Window, Interactor
renderer = vtk.vtkRenderer()
renderer.AddVolume(volume)
renderer.SetBackground(0.1, 0.1, 0.1)  # dark background
renderer.ResetCamera()

renderWindow = vtk.vtkRenderWindow()
renderWindow.AddRenderer(renderer)
renderWindow.SetSize(800, 800)

interactor = vtk.vtkRenderWindowInteractor()
interactor.SetRenderWindow(renderWindow)

# Step 7: Start interaction
renderWindow.Render()
interactor.Start()
```
